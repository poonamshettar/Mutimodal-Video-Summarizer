```python
video1 = cv2.VideoCapture('D:\CEVI\\videos\mixkit-small-pink-flowers-1186-medium.mp4')
if (video1.isOpened()== False):
	print("Error opening video file")
while(video1.isOpened()):
	ret, frame = video1.read()
	if ret == True:
		cv2.imshow("video",frame)
		if cv2.waitKey(25) & 0xFF == ord('q'):
			break
		
	else:
		break
video1.release()
cv2.destroyAllWindows()
```

```python
from moviepy.editor import VideoFileClip
import matplotlib.pyplot as plt
clip1 = VideoFileClip("x.mp4").subclip(0, 14)
clip2 = VideoFileClip("y.mp4").subclip(0, 14)
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].imshow(clip1.get_frame(0))
axes[0].axis('off')
axes[0].set_title('Video 1')
axes[1].imshow(clip2.get_frame(0))
axes[1].axis('off')
axes[1].set_title('Video 2')
plt.show()

```

```python
import cv2

vidcap = cv2.VideoCapture('y.mp4')
count = 0
success = True
fps = int(vidcap.get(cv2.CAP_PROP_FPS))

while success:
    success,image = vidcap.read()
    print('read a new frame:',success)
    if count%(2*fps) == 0 :
         cv2.imwrite('frame%d.jpg'%count,image)
         print('successfully written 10th frame')
    count+=1
```

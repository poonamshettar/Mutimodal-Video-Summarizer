```python
from scipy.signal import convolve2d
import numpy as np
gaussian = (1 / 16.0) * np.array([[1., 2., 1.],
                                  [2., 4., 2.],
                                  [1., 2., 1.]])
def multi_convolver(image, kernel, iterations):
    for i in range(iterations):
        image = convolve2d(image, kernel, 'same', boundary = 'fill',
                           fillvalue = 0)
    return image
img1=multi_convolver(img,gaussian,15)
plt.imshow(img1)
np.shape(img1)
frame=cv2.imread('video_frames_2\\frame0.jpg')
```

remove garbage/unwanted/meaningless text

```python
from nltk import pos_tag
from spellchecker import SpellChecker
import re
import nltk
# nltk.download('stopwords')
spell = SpellChecker()
# stopwords = []
word_list_pos_tagged = [ pos_tag(word) for word in word_list]
print(word_list_pos_tagged)
final = []

for words in word_list_pos_tagged:
    sentence_final = []  
    for word, pos in words:
        if (
            spell.correction(word) == word
            and pos in ('NN', 'DT', 'CC', 'EX', 'FW', 'IN', 'JJ', 'JJR', 'JJS', 'NNP', 'LS', 'MD', 'NNS', 'NNPS', 'PDT', 'POS', 'PRP', 'PRP$', 'RB', 'RBR', 'RBS', 'RP', 'TO', 'UH', 'VB', 'VBD', 'VBN', 'VBP', 'VBZ', 'WDT', 'WP', 'WP$', 'WRB')
        ):
            word=re.sub(r'[^A-Za-z0-9\s]', '', word)
            sentence_final.append(word) 
    sentence_final=" ".join(sentence_final) 
    final.append(sentence_final)  
print(final)
```

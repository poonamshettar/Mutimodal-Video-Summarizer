image to text using pytesseract

```python
import cv2
import pytesseract as p
from nltk.corpus import wordnet
x=[]
p.pytesseract.tesseract_cmd = r'C:\Users\Lenovo\AppData\Local\Programs\Tesseract-OCR\tesseract.exe'
prev=0
for j in range(38):
    img_path = r"video_frames_1\\frame{}.jpg".format(j)
    image = cv2.imread(img_path)
    text=p.image_to_string(image)
    if(prev != text):
      x.append(text)
      print(text)
    prev=text
```

remove newline charecter

```python
print(x)
new_list = [
    item.replace('\n', ' ')
    for item in x
]
print(new_list)
' '.join(new_list)
```

tokenize into words

```python
import nltk
from nltk import word_tokenize
word_list=[]
for words in new_list:
    word=word_tokenize(words)
    word_list.append(word)
print(word_list)
```

remove garbage/unwanted/meaningless text

```python
from nltk import pos_tag
from spellchecker import SpellChecker
import re
import nltk
# nltk.download('stopwords')
spell = SpellChecker()
# stopwords = []
word_list_pos_tagged = [ pos_tag(word) for word in word_list]
print(word_list_pos_tagged)
final = []

for words in word_list_pos_tagged:
    sentence_final = []  
    for word, pos in words:
        if (
            spell.correction(word) == word
            and pos in ('NN', 'DT', 'CC', 'EX', 'FW', 'IN', 'JJ', 'JJR', 'JJS', 'NNP', 'LS', 'MD', 'NNS', 'NNPS', 'PDT', 'POS', 'PRP', 'PRP$', 'RB', 'RBR', 'RBS', 'RP', 'TO', 'UH', 'VB', 'VBD', 'VBN', 'VBP', 'VBZ', 'WDT', 'WP', 'WP$', 'WRB')
        ):
            word=re.sub(r'[^A-Za-z0-9\s]', '', word)
            sentence_final.append(word) 
    sentence_final=" ".join(sentence_final) 
    final.append(sentence_final)  
print(final)
```

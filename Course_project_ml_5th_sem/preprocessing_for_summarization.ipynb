```python
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df=pd.read_excel('data\\news.xlsx')
```

```python
df.head()
```

```python
Text=df['Short']
Summary=df['Headline']
```

```python
Text[0]
```

```python
Text
type(Text)
Text=Text.tolist()
print(Text[:2])
type(Text)
```

```python
stop_words = stopwords.words("english")
vectorizer = TfidfVectorizer()
```

```python
for i in range(len(Text)):
    sentences=sent_tokenize(Text[i])
    k = [p for m in sentences for p in m.split(' ') if p not in stop_words]
    sentences = sent_tokenize(' '.join(k))
    vectors = vectorizer.fit_transform(sentences)
    feature_names = vectorizer.get_feature_names_out()
    dense = vectors.todense()
    dict={}
    for k in range(len(sentences)):
        dict[sentences[k]]=dense[k].sum()
    sent = []
    ave=(sum(dict.values())/len(dict.values()))
    for j in dict.items():
        if j[1]>=ave:
            sent.append(j[0])
    sent=' '.join(sent)
    Text[i]=sent
```

```python
def evaluation2():
    from rouge import Rouge
    rouge=Rouge()
    scores=rouge.get_scores(Summary[0],Text[0],avg=True)
    print("Rouge-Lsum Precision:", scores["rouge-l"]["p"])
    print("Rouge-Lsum Recall:", scores["rouge-l"]["r"])
    print("Rouge-Lsum F1 Score:", scores["rouge-l"]["f"])
evaluation2()
```

```python
def evaluation1():
    from rouge_score import rouge_scorer
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    scores = scorer.score(Text[2],Summary[2])
    for key in scores:
        print(f'{key}: {scores[key]}')
evaluation1()
```
